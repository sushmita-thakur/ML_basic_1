Ans->>A parameter is a configuration variable internal to the model whose value is estimated from the data during training. Parameters define how the model makes predictions. Examples include weights in linear regression or coefficients in logistic regression.

4. How does loss value help in determining whether the model is good or not?
Ans->>The loss value quantifies the error between predicted and actual values. A lower loss indicates better model performance on the training data, but must be evaluated on unseen data to avoid overfitting.

Ans->>An optimizer is an algorithm that minimizes the loss function by updating model parameters.
Types:

Gradient Descent: Updates parameters in the direction of steepest descent.

Adam: Adaptive learning rate method combining RMSprop and Momentum.

SGD: Stochastic Gradient Descent updates parameters using a subset of data.
